```python
# This program makes the heatmap for a corresponding data file.
# The center location is automatically calculated by the library and need not be specified by the user.
# This is a testbed with data being taken from https://earthquake.usgs.gov/earthquakes/search/ 
# The data corresponds to earthquakes recorded from 2020-03-07 00:00:00 to 2020-04-06 23:59:59.
# Please free to experiment with the max_intensity and point_radius values.
# A more extensive dataset can be recorded from the aforementioned website.
```

```python
import gmaps
import gmaps.datasets
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('./../query.csv') # Loading the earthquake data
data.head()
```

```python
locations = data[['latitude','longitude']] # defining the locations of earthquakes recorded
magnitude = data['mag'] # magnitude of earthquake recorded
fig = gmaps.figure(center=(0,0),zoom_level=2) # making map with center set to (0.0,0.0) and zoom level to 2
heatmap_layer = gmaps.heatmap_layer(locations, weights = magnitude) # creating the datapoints to map with weights
# corresponding to the magnitude of the earthquake.
fig.add_layer(heatmap_layer) # adding the layer to the map
fig
```

```python
# max_intensity is used to prevent dissipating.
heatmap_layer.max_intensity = 50 # ratio of highest to lowest something. Experiment with this to see the above map change.
heatmap_layer.point_radius = 3 # zone of influence of each point. Lower the value greater the point differentiation
```
